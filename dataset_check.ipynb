```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import random as rd
```

```python
dataset = pd.read_csv("dataset.csv")
dataset.head()
```

```python
augmented_dataset = pd.read_csv("augmented_dataset.csv")
augmented_dataset.head()
```

```python
NUM_IMAGES = 5
plt.figure(figsize=(30, 18))  # Increased figure size for bigger images
for i in range(NUM_IMAGES):
    img_path, x, y = dataset.iloc[rd.randint(0, len(dataset))]
    
    # mirror x and y
    x = 1 - x
    y = 1 - y

    # Image
    img = plt.imread(img_path)
    
    plt.subplot(NUM_IMAGES, 2, 2*i+1)
    plt.imshow(img)
    plt.title(f"{img_path}", fontsize=14)  # Increased font size

    # Target (x, y) in a 1×1 square
    ax = plt.subplot(NUM_IMAGES, 2, 2*i+2)
    ax.plot(x, y, 'ro', markersize=10)  # Increased marker size

    # keep the limits fixed at [0, 1] even after enforcing square aspect
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal', adjustable='box')

    # optional: show only the 0 and 1 ticks
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])

    ax.set_title(f"Target {x:.2f} {y:.2f}", fontsize=14)  # Increased font size

plt.tight_layout(pad=4.0)  # Increased padding for better spacing
plt.show()
```

```python
NUM_IMAGES = 5
plt.figure(figsize=(30, 18))  # Increased figure size for bigger images
for i in range(NUM_IMAGES):
    img_path, x, y = augmented_dataset.iloc[rd.randint(0, len(augmented_dataset))]
    
    # mirror x and y
    x = 1 - x
    y = 1 - y

    # Image - convert to grayscale
    img = plt.imread(img_path)
    if len(img.shape) == 3:  # If RGB image
        img_gray = 0.299 * img[:,:,0] + 0.587 * img[:,:,1] + 0.114 * img[:,:,2]
    else:
        img_gray = img
    
    plt.subplot(NUM_IMAGES, 2, 2*i+1)
    plt.imshow(img_gray, cmap='gray')
    plt.title(f"{img_path}", fontsize=14)  # Increased font size

    # Target (x, y) in a 1×1 square
    ax = plt.subplot(NUM_IMAGES, 2, 2*i+2)
    ax.plot(x, y, 'ro', markersize=10)  # Increased marker size

    # keep the limits fixed at [0, 1] even after enforcing square aspect
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.set_aspect('equal', adjustable='box')

    # optional: show only the 0 and 1 ticks
    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])

    ax.set_title(f"Target {x:.2f} {y:.2f}", fontsize=14)  # Increased font size

plt.tight_layout(pad=4.0)  # Increased padding for better spacing
plt.show()
```
